```r
# 홀수 짝수 구분

a <- 8
class(a)
# 8은 짝수 입니다.

if(a%%2 == 0){
    cat(a, "은 짝수 입니다.", sep = "")
}else{
    cat(a, "는 짝수가 아닙니다.", sep = "")
}
```

```r
# 홀짝수 구분
a <- c(1,2,3)
ifelse(a %% 2 == 0, "짝수","홀수")
```

```r
# 홀짝수 구분
a <- c(1,2,3)
ifelse(a %% 2 == 0, paste(a,"는 짝수"), paste(a,"는 홀수"))
```

```r
# excel.csv를 읽어서 excel이라는 벡터변수에 할당하기

excel <- read.csv("./Data//excel.csv"
                  )
head(excel)

tail(excel)

str(excel)

summary(excel)

```

```r
# q1 컬럼에서 Data가 3 보다 크거나 같으면 Data의 제곱근을 구하고 아니면 원래 Data 출력하기

q1 <- excel$q1
ifelse(q1 >= 3 , sqrt(q1), q1)
```

```r
# 위 결과를 q6라는 컬럼을 만들어 DataSet에 추가하기

excel$q6 <- ifelse(q1 >= 3 , sqrt(q1), q1)
head(excel)
```
