```r
# excel file에서 q3컬럼의 홀수 합과 짝수의 합 구하기
excel <- read.csv("./Data//excel.csv")

head(excel)

oddq3 <- 0
evenq3 <- 0

for(i in excel$q3){
    if(i %% 2 == 0){
        oddq3 <- oddq3 + i
    }else{
        evenq3 <- evenq3 + i
    }
    
}
cat(oddq3,"\n")
cat(evenq3)

q3 <- excel$q3

tapply(q3,ifelse(q3 %% 2 == 0, "짝수" , "홀수"),sum)

sum(ifelse(q3 %% 2 == 0, q3, 0))
sum(ifelse(q3 %% 2 == 1, q3, 0))
```

```r
# excel file에서 q3 컬럼 데이터중 3의 배수도 되고 4의 배수도 되는 수의 합계를 구하시오


sum(ifelse(q3 %% 3 == 0 & q3 %% 4 == 0 , q3,0))
```

```r
# 구구단 중 짝수단 과 짝수 곱해지는 수만 출력하기

for(i in 1:9){
    for(dan in 2:9){
        cat(i, "X", dan , "=" , i * dan,'\n')
    }
    cat('--------------\n')
}
```

```r
# 2개의 벡터 자료를 함께 출력하기
# paste 벡터를 새롭게 만드는 함수
score <- c(85, 95, 89)
name <- c('유비', '관우', '장비')

for(i in 1:length(name)){
    cat(name[i], ":" , score[i],"\n")
}


cat(paste(name, ":", score), sep = "\n")
```

# while

```r
i <- 0
while(T){
    i = i + 1
    print(i)
    if(i > 9){
        break
    }
}
```

# repeat

```r
i <- 0

repeat{
    i <- i + 1
    if(i > 10){
        break
    }
    print(i)
}
```
